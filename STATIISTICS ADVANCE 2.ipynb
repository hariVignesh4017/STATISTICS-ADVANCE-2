# 4. What does a P-value represent in hypothesis testing

# 11. What is the difference between a Z-test and a T-test
A Z-test and a T-test are both used in hypothesis testing to compare means, but they differ based on sample size and whether the population variance is known.

A Z-test is used when the sample size is large (n â‰¥ 30) and the population variance (ÏƒÂ²) is known. It assumes that the data follows a normal distribution. Since larger samples tend to follow the Central Limit Theorem, the normal distribution provides accurate results.

On the other hand, a T-test is used when the sample size is small (n < 30) and the population variance is unknown. Instead of using the normal distribution, it follows the T-distribution, which has heavier tails. This accounts for the additional variability seen in smaller samples.

The T-test is more conservative because it widens confidence intervals, reducing the chance of incorrect conclusions when sample sizes are small. However, as the sample size increases, the T-distribution gradually approaches the normal distribution, making both tests nearly identical for large samples.

# 12. What is the T-test, and how is it used in hypothesis testing
A T-test is a statistical hypothesis test used to compare the means of one or two groups and determine if they are significantly different from each other. It is used when the sample size is small (n < 30) or when the population variance is unknown. Instead of the normal distribution, it follows the T-distribution, which has heavier tails to account for more variability in small samples.

The test calculates how many standard errors the sample mean is away from the population mean (or difference between two sample means).

Compare with Critical Value or p-Value:
If the p-value < significance level (0.05), reject the null hypothesis, meaning there is a significant difference.
If the p-value â‰¥ 0.05, fail to reject the null hypothesis (no significant difference).

# 14. What is a confidence interval, and how is it used to interpret statistical results
A confidence interval (CI) is a range of values, derived from sample data, that likely contains the true population parameter (such as a mean or proportion) with a certain level of confidence. Instead of providing a single estimate, it gives a range where the true value is expected to fall.

For example, a 95% confidence interval means that if we were to repeat the experiment many times, 95% of the intervals calculated would contain the true population parameter.

How is a Confidence Interval Used in Statistical Interpretation?

Estimate Population Parameters
Confidence intervals provide a range of plausible values for an unknown population parameter based on a sample statistic.

Assess Statistical Significance
If a confidence interval for a difference excludes zero, it suggests a significant difference between groups.
If a confidence interval for a mean excludes a certain value (e.g., 100), we can infer that the true mean is likely different from that value.

Compare with Hypothesis Testing
A confidence interval and hypothesis test often lead to the same conclusion.
If a 95% CI does not include the null hypothesis value, we reject the null hypothesis at Î± = 0.05 significance level.

Width of the Interval Indicates Precision
Narrow intervals indicate more precise estimates (less uncertainty).
Wide intervals suggest more variability in the data.
Larger sample sizes lead to narrower confidence intervals due to reduced variability.

# 15. What is the margin of error, and how does it affect the confidence interval
The margin of error (MOE) is the amount of uncertainty or variability allowed in a statistical estimate. It represents how much the sample estimate (e.g., mean, proportion) could differ from the true population parameter due to random sampling variability.
In confidence intervals, the margin of error is the distance from the sample statistic to the upper and lower bounds of the interval. It ensures that the confidence interval accounts for potential variations in different samples.
For example, if a survey reports a polling result of 55% with a margin of error of Â±3%, the actual percentage in the population is expected to be between 52% and 58%.

How Does the Margin of Error Affect the Confidence Interval?
Determines Interval Width
A larger margin of error results in a wider confidence interval, meaning less precision but more certainty that it contains the true population value.
A smaller margin of error results in a narrower confidence interval, meaning greater precision but a higher risk of missing the true value.

Depends on Sample Size
Larger samples reduce the margin of error, leading to a narrower confidence interval.
Smaller samples increase the margin of error, leading to a wider confidence interval.

Influenced by Confidence Level
A higher confidence level (e.g., 99%) increases the margin of error, making the confidence interval wider.
A lower confidence level (e.g., 90%) decreases the margin of error, making the confidence interval narrower.

Impacts Statistical Decision-Making
A large margin of error means more uncertainty, making it harder to draw strong conclusions.
A small margin of error increases confidence in the estimate but may require more data.

# 20.  What is an ANOVA test, and what are its assumptions
ANOVA (Analysis of Variance) is a statistical test used to compare the means of multiple groups to determine if there is a significant difference between them. Instead of performing multiple t-tests (which increase error rates), ANOVA analyzes variances to draw conclusions about group means.
One-Way ANOVA: Compares the means of one factor (e.g., comparing exam scores across three different schools).
Two-Way ANOVA: Compares the means of two independent factors (e.g., testing if both study method and gender affect test scores).
ANOVA uses the F-distribution to determine whether the differences between group means are statistically significant.

Assumptions of ANOVA
For ANOVA to be valid, these assumptions must be met:

Normality
The dependent variable should be normally distributed within each group.
Test: Shapiro-Wilk test, Kolmogorov-Smirnov test, or Q-Q plots.

Independence
Observations should be independent of each other.
This means no paired samples or repeated measures.

Homogeneity of Variance (Homoscedasticity)
All groups should have similar variances.
Test: Leveneâ€™s test or Bartlettâ€™s test.
If these assumptions are violated, alternative tests like Kruskal-Wallis (for non-normal data) or Welchâ€™s ANOVA (for unequal variances) can be used.

```python
# 1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

# Sample data
sample = [52, 48, 50, 49, 51, 53, 47, 50, 49, 51]  # Example sample values
population_mean = 50  # Known population mean
population_std = 2  # Known population standard deviation
alpha = 0.05  # Significance level

# Calculate sample mean and sample size
sample_mean = np.mean(sample)
sample_size = len(sample)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference between the sample and population mean.")
```

```python
# 3  Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05, tail='two'):
    """
    Perform a one-sample Z-test.
    
    Parameters:
    sample : list or array - Sample data
    pop_mean : float - Known population mean
    pop_std : float - Known population standard deviation
    alpha : float - Significance level (default=0.05)
    tail : str - 'two' for two-tailed test, 'left' for left-tailed, 'right' for right-tailed
    
    Returns:
    Z-score, p-value, and interpretation of results
    """
    # Compute sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value
    if tail == 'two':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif tail == 'left':
        p_value = stats.norm.cdf(z_score)
    elif tail == 'right':
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("Invalid tail type. Use 'two', 'left', or 'right'.")

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis (H0): The sample mean is significantly different from the population mean."
    else:
        result = "Fail to reject the null hypothesis (H0): No significant difference detected."

    return z_score, p_value, result

# Example Usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample
population_mean = 100
population_std = 15
alpha = 0.05

# Perform Z-test
z_score, p_value, interpretation = one_sample_z_test(sample_data, population_mean, population_std, alpha, tail='two')

# Print Results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
# 4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate Sample Data
np.random.seed(42)
sample_size = 30
population_mean = 100  # Population mean (H0)
population_std = 15  # Known population standard deviation
alpha = 0.05  # Significance level

sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)  # Simulated sample

# Step 2: Compute Z-Score
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 3: Compute P-Value (Two-Tailed Test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 4: Compute Critical Z-Values for Two-Tailed Test
z_critical = stats.norm.ppf(1 - alpha / 2)  # Positive critical value
z_critical_neg = -z_critical  # Negative critical value

# Step 5: Interpretation
if abs(z_score) > z_critical:
    result = "Reject the null hypothesis (H0): The sample mean is significantly different."
else:
    result = "Fail to reject the null hypothesis (H0): No significant difference detected."

# Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Z-Critical Values: Â±{z_critical:.4f}")
print(f"Interpretation: {result}")

# Step 6: Visualization of Decision Region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal Distribution")

# Shade rejection regions
plt.fill_between(x, y, where=(x < z_critical_neg), color='red', alpha=0.3, label="Rejection Region (Left)")
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3, label="Rejection Region (Right)")

# Plot critical values
plt.axvline(z_critical, color='red', linestyle='dashed', label=f"Z = {z_critical:.2f}")
plt.axvline(z_critical_neg, color='red', linestyle='dashed')

# Plot observed Z-score
plt.axvline(z_score, color='blue', linestyle='solid', label=f"Observed Z = {z_score:.2f}")

# Labels and legend
plt.title("Two-Tailed Z-Test: Decision Regions")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
# 5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.
    
    Parameters:
    mu0 : float  -> Null hypothesis population mean
    mu1 : float  -> Alternative hypothesis population mean
    sigma : float -> Population standard deviation
    n : int -> Sample size
    alpha : float -> Significance level (default 0.05)
    """
    
    # Calculate standard error
    se = sigma / np.sqrt(n)

    # Critical Z-value for alpha (two-tailed)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Calculate critical value (boundary for rejecting H0)
    critical_value_right = mu0 + z_critical * se
    critical_value_left = mu0 - z_critical * se

    # Calculate Type 2 Error (Beta)
    beta = stats.norm.cdf(critical_value_right, loc=mu1, scale=se) - stats.norm.cdf(critical_value_left, loc=mu1, scale=se)
    power = 1 - beta  # Power of the test

    # Create X values for distribution plots
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)
    
    # Normal distributions for H0 and H1
    y_h0 = stats.norm.pdf(x, mu0, se)
    y_h1 = stats.norm.pdf(x, mu1, se)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label="Null Distribution (H0)", color="blue")
    plt.plot(x, y_h1, label="Alternative Distribution (H1)", color="green")

    # Shade Type 1 Error (Alpha - rejecting H0 when it's true)
    plt.fill_between(x, y_h0, where=(x >= critical_value_right), color='red', alpha=0.3, label="Type 1 Error (Alpha)")
    plt.fill_between(x, y_h0, where=(x <= critical_value_left), color='red', alpha=0.3)

    # Shade Type 2 Error (Beta - failing to reject H0 when H1 is true)
    plt.fill_between(x, y_h1, where=(x > critical_value_left) & (x < critical_value_right), color='orange', alpha=0.3, label="Type 2 Error (Beta)")

    # Add vertical lines for critical values
    plt.axvline(critical_value_right, color='red', linestyle='dashed', label=f"Critical Value (+{z_critical:.2f})")
    plt.axvline(critical_value_left, color='red', linestyle='dashed', label=f"Critical Value (-{z_critical:.2f})")

    # Labels and legend
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid()
    plt.show()

    # Print results
    print(f"Critical Values: {critical_value_left:.2f}, {critical_value_right:.2f}")
    print(f"Type 1 Error (Alpha): {alpha:.2f}")
    print(f"Type 2 Error (Beta): {beta:.4f}")
    print(f"Power of the Test: {power:.4f}")

# Example usage
visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

```python
# 6 Write a Python program to perform an independent T-test and interpret the results
import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True):
    """
    Performs an independent (two-sample) t-test and interprets the results.
    
    Parameters:
    sample1 : list or array - Data for group 1
    sample2 : list or array - Data for group 2
    alpha : float - Significance level (default 0.05)
    equal_var : bool - Assume equal variance (True) or Welchâ€™s t-test (False)
    
    Returns:
    t-statistic, p-value, and result interpretation.
    """
    # Perform T-test
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var)

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis (H0): The two groups have significantly different means."
    else:
        result = "Fail to reject the null hypothesis (H0): No significant difference between the group means."

    return t_stat, p_value, result

# Example: Generate random samples for two groups
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: Mean = 100, Std = 15
group2 = np.random.normal(loc=110, scale=15, size=30)  # Group 2: Mean = 110, Std = 15

# Perform the independent T-test
t_stat, p_value, interpretation = independent_t_test(group1, group2, alpha=0.05, equal_var=True)

# Print Results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
# 7 Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(pre, post, alpha=0.05):
    """
    Performs a paired (dependent) t-test and visualizes the results.

    Parameters:
    pre : list or array - Before treatment (Pre-test)
    post : list or array - After treatment (Post-test)
    alpha : float - Significance level (default 0.05)

    Returns:
    t-statistic, p-value, and result interpretation.
    """
    # Perform Paired T-Test
    t_stat, p_value = stats.ttest_rel(pre, post)

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis (H0): There is a significant difference before and after treatment."
    else:
        result = "Fail to reject the null hypothesis (H0): No significant difference detected."

    # Visualization
    plt.figure(figsize=(8, 5))
    
    # Scatter plot to show pre vs. post comparison
    plt.scatter(pre, post, color='blue', alpha=0.6, label="Pre vs. Post")
    plt.plot([min(pre), max(pre)], [min(pre), max(pre)], linestyle="--", color="red", label="y = x (No Change)")
    
    plt.xlabel("Pre-Test Scores")
    plt.ylabel("Post-Test Scores")
    plt.title("Paired Sample T-Test: Pre vs. Post Scores")
    plt.legend()
    plt.grid()
    plt.show()

    return t_stat, p_value, result

# Example: Generate paired sample data
np.random.seed(42)
pre_scores = np.random.normal(loc=75, scale=10, size=20)  # Before treatment
post_scores = pre_scores + np.random.normal(loc=5, scale=5, size=20)  # After treatment (improvement)

# Perform Paired T-Test
t_stat, p_value, interpretation = paired_t_test(pre_scores, post_scores)

# Print Results
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
# 8 Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
import scipy.stats as stats

def simulate_data(n=25, mu=100, sigma=15):
    """
    Generates a random sample from a normal distribution.
    Parameters:
        n : int -> Sample size
        mu : float -> Population mean
        sigma : float -> Population standard deviation
    Returns:
        sample (array)
    """
    np.random.seed(42)
    return np.random.normal(loc=mu, scale=sigma, size=n)

def perform_z_test(sample, pop_mean=100, pop_std=15):
    """
    Performs a one-sample Z-test.
    Parameters:
        sample : array -> Sample data
        pop_mean : float -> Population mean (H0)
        pop_std : float -> Population standard deviation
    Returns:
        Z-statistic, P-value
    """
    sample_mean = np.mean(sample)
    n = len(sample)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed
    return z_stat, p_value

def perform_t_test(sample, pop_mean=100):
    """
    Performs a one-sample T-test.
    Parameters:
        sample : array -> Sample data
        pop_mean : float -> Population mean (H0)
    Returns:
        T-statistic, P-value
    """
    t_stat, p_value = stats.ttest_1samp(sample, pop_mean)
    return t_stat, p_value

# Step 1: Simulate Data
sample = simulate_data(n=25, mu=100, sigma=15)

# Step 2: Perform Z-Test (Assuming population std dev is known)
z_stat, z_p_value = perform_z_test(sample, pop_mean=100, pop_std=15)

# Step 3: Perform T-Test (When population std dev is unknown)
t_stat, t_p_value = perform_t_test(sample, pop_mean=100)

# Step 4: Print Results
print("=== Z-Test Results ===")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {z_p_value:.4f}")

print("\n=== T-Test Results ===")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {t_p_value:.4f}")

# Step 5: Interpretation
if z_p_value < 0.05:
    print("\nZ-Test: Reject H0 (Significant difference detected)")
else:
    print("\nZ-Test: Fail to reject H0 (No significant difference)")

if t_p_value < 0.05:
    print("T-Test: Reject H0 (Significant difference detected)")
else:
    print("T-Test: Fail to reject H0 (No significant difference)")
```

```python
# 9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Computes the confidence interval for a sample mean.

    Parameters:
        sample : array -> Sample data
        confidence : float -> Confidence level (default 0.95 for 95% CI)

    Returns:
        (lower_bound, upper_bound) -> Confidence interval range
    """
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    n = len(sample)

    # Calculate the margin of error
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # T-score for given confidence level
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example: Generate a random sample
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Sample from normal distribution

# Compute the 95% Confidence Interval
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

# Print Results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 10  Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """
    Computes the margin of error for a sample mean.
    
    Parameters:
        sample : array -> Sample data
        confidence : float -> Confidence level (default 0.95 for 95%)
    
    Returns:
        Margin of Error (float)
    """
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)
    n = len(sample)
    
    # Get the critical T-score
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    # Compute Margin of Error
    moe = t_critical * (sample_std / np.sqrt(n))
    
    return moe

# Example: Generate random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Sample from normal distribution

# Compute Margin of Error
moe = margin_of_error(sample_data, confidence=0.95)

# Print Results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Margin of Error (95% Confidence): Â±{moe:.2f}")
```

```python
# 11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, false_positive):
    # Total probability of a positive test (evidence)
    evidence = likelihood * prior + false_positive * (1 - prior)
    
    # Posterior probability
    posterior = (likelihood * prior) / evidence
    return posterior

# Given values
prior = 0.01  # 1% of the population has the disease
likelihood = 0.9  # True positive rate
false_positive = 0.05  # False positive rate

# Compute posterior probability
posterior = bayes_theorem(prior, likelihood, false_positive)

print(f"Probability of having the disease given a positive test: {posterior:.2%}")
```

```python
# 13  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def expected_frequencies(observed):
    observed = np.array(observed)
    
    # Compute row and column totals
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    
    # Compute grand total
    grand_total = observed.sum()
    
    # Compute expected frequencies
    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example observed data (Contingency Table)
observed_data = [[30, 20],  # Coffee preference
                 [10, 40]]  # Tea preference

# Calculate expected frequencies
expected_values = expected_frequencies(observed_data)

# Display results
print("Observed Data:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(np.round(expected_values, 2))
```

```python
# 16 Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

# Generate two random samples (assume normal distribution)
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10
sample2 = np.random.normal(loc=55, scale=15, size=30)  # Mean=55, Std Dev=15

# Compute sample variances
var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F_stat = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F â‰¥ 1

# Compute p-value (two-tailed test)
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * (1 - f.cdf(F_stat, df1, df2))  # Two-tailed test

# Display results
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {F_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
# 19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):
    """
    Check normality, independence, and equal variance assumptions for ANOVA.
    
    Parameters:
        groups (list of arrays): A list where each element is a group (numpy array).
    """
    
    print("### Checking ANOVA Assumptions ###\n")

    # Normality test (Shapiro-Wilk test)
    print("1. Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"   Group {i+1}: W-stat={stat:.3f}, p-value={p:.4f}")
        if p < 0.05:
            print("   âŒ Likely not normal (reject H0)")
        else:
            print("   âœ… Likely normal (fail to reject H0)")
    
    # Independence: This is usually ensured by study design (random sampling).
    print("\n2. Independence: Should be ensured by study design.")

    # Equal variance test (Leveneâ€™s test)
    print("\n3. Equal Variance Test (Leveneâ€™s Test):")
    stat, p = stats.levene(*groups)
    print(f"   Leveneâ€™s Test: W-stat={stat:.3f}, p-value={p:.4f}")
    if p < 0.05:
        print("   âŒ Variances are significantly different (reject H0)")
    else:
        print("   âœ… Variances are similar (fail to reject H0)")

    # Visualizing normality with histograms
    plt.figure(figsize=(8, 5))
    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=f"Group {i+1}", alpha=0.6)
    
    plt.title("Histogram of Groups (Check Normality)")
    plt.xlabel("Values")
    plt.legend()
    plt.show()

# Example Usage
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

# Check ANOVA assumptions
check_anova_assumptions([group1, group2, group3])
```

```python
# 20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Generate sample data
np.random.seed(42)
data = {
    'Diet': np.repeat(['A', 'B'], 30),
    'Exercise': np.tile(['Low', 'High'], 30),
    'Weight_Loss': np.concatenate([
        np.random.normal(loc=5, scale=1.5, size=15),  # Diet A, Low Exercise
        np.random.normal(loc=6, scale=1.5, size=15),  # Diet A, High Exercise
        np.random.normal(loc=7, scale=1.5, size=15),  # Diet B, Low Exercise
        np.random.normal(loc=9, scale=1.5, size=15)   # Diet B, High Exercise
    ])
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Perform Two-Way ANOVA
model = ols('Weight_Loss ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)', data=df).fit()
anova_results = anova_lm(model)

# Display results
print("\nTwo-Way ANOVA Results:")
print(anova_results)

# Interpretation
alpha = 0.05
for factor in ['C(Diet)', 'C(Exercise)', 'C(Diet):C(Exercise)']:
    p_val = anova_results.loc[factor, 'PR(>F)']
    if p_val < alpha:
        print(f"âœ… {factor} has a significant effect (p = {p_val:.4f})")
    else:
        print(f"âŒ {factor} has no significant effect (p = {p_val:.4f})")
```

```python
# 21  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(df1, df2):
    """
    Plots the F-distribution for given degrees of freedom (df1, df2).
    
    Parameters:
        df1 (int): Degrees of freedom for the numerator.
        df2 (int): Degrees of freedom for the denominator.
    """
    x = np.linspace(0, 5, 500)  # F-values
    y = stats.f.pdf(x, df1, df2)  # Probability Density Function (PDF)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, alpha=0.3, color='blue')  # Fill area under the curve

    # Critical region for a 95% confidence level (alpha=0.05)
    alpha = 0.05
    critical_value = stats.f.ppf(1 - alpha, df1, df2)
    plt.axvline(critical_value, color='red', linestyle='dashed', label=f'Critical Value = {critical_value:.2f}')

    # Labels and title
    plt.title("F-Distribution and Critical Region")
    plt.xlabel("F-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example Usage
plot_f_distribution(df1=5, df2=10)
```

```python
# 22 Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group mean
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
# Set random seed for reproducibility
np.random.seed(42)

# Create three groups with different means
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60

# Convert to DataFrame for visualization
df = pd.DataFrame({
    'Value': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1']*30 + ['Group 2']*30 + ['Group 3']*30
})
# Perform ANOVA test
F_stat, p_value = f_oneway(group1, group2, group3)

# Display results
print(f"F-statistic: {F_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")
    # Set style
sns.set(style="whitegrid")

# Create a boxplot to visualize group distributions
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Value', data=df, palette="Set2")

# Add title and labels
plt.title("One-Way ANOVA: Comparison of Group Means")
plt.xlabel("Group")
plt.ylabel("Value")

# Show the plot
plt.show()
```

```python
# 24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# Simulate data from a normal distribution
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Population mean=50, std=10
sample_size = len(sample)

# Hypothesis
pop_variance = 100  # Assumed population variance (ÏƒÂ²)
sample_variance = np.var(sample, ddof=1)  # Sample variance (sÂ²)

# Chi-Square Test Statistic
chi_square_stat = (sample_size - 1) * sample_variance / pop_variance

# Critical values & p-value
alpha = 0.05
df = sample_size - 1  # Degrees of freedom
chi_crit_low = stats.chi2.ppf(alpha / 2, df)
chi_crit_high = stats.chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Display results
print(f"Chi-Square Statistic: {chi_square_stat:.2f}")
print(f"Critical Values: ({chi_crit_low:.2f}, {chi_crit_high:.2f})")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if chi_crit_low < chi_square_stat < chi_crit_high:
    print("Fail to reject the null hypothesis: The variance is not significantly different.")
else:
    print("Reject the null hypothesis: The variance is significantly different.")

```

```python
# 25  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, size_a, success_b, size_b, alpha=0.05):
    """
    Perform a Z-test for two population proportions.
    
    Parameters:
        success_a (int): Number of successes in group A.
        size_a (int): Sample size of group A.
        success_b (int): Number of successes in group B.
        size_b (int): Sample size of group B.
        alpha (float): Significance level (default=0.05).
        
    Returns:
        Prints Z-statistic, P-value, and interpretation.
    """
    # Calculate sample proportions
    p1 = success_a / size_a
    p2 = success_b / size_b

    # Pooled proportion
    p_pooled = (success_a + success_b) / (size_a + size_b)

    # Standard error
    se = np.sqrt(p_pooled * (1 - p_pooled) * (1/size_a + 1/size_b))

    # Z-score
    z_stat = (p1 - p2) / se

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Output results
    print(f"Z-Statistic: {z_stat:.2f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in proportions.")

# Example usage
z_test_proportions(success_a=50, size_a=200, success_b=65, size_b=220)
```

```python
# 26  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate two random datasets
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15 (higher variance)

# Compute sample variances
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# Perform F-test (larger variance in numerator)
F_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(group1) - 1
df2 = len(group2) - 1
p_value = 2 * (1 - stats.f.cdf(F_stat, df1, df2))  # Two-tailed test

# Display results
print(f"F-Statistic: {F_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")

# Visualization
plt.figure(figsize=(8,5))
sns.histplot(group1, kde=True, label="Group 1 (Lower Variance)", color='blue', alpha=0.6)
sns.histplot(group2, kde=True, label="Group 2 (Higher Variance)", color='red', alpha=0.6)
plt.title("Distribution Comparison (F-Test)")
plt.legend()
plt.show()
```

```python
# 27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats

# Simulated observed frequencies
observed = np.array([40, 35, 25])  # Example: Counts of categories

# Expected frequencies (assume uniform distribution)
expected = np.array([sum(observed) / len(observed)] * len(observed))

# Perform Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Display results
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is significantly different from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the expected distribution.")
```
